# Modules

```python
import numpy as np
import plotly.graph_objects as go
from scipy.stats import levy_stable
```

# Par√°metros

```python
m = 3.0 
beta = 0
x = np.linspace(-10, 10, 1000)  
alpha_values = [0.5, 1.0, 1.5, 2.0] 
fig = go.Figure()
```

# Curves Generator

```python
for alpha in alpha_values:
    y = levy_stable.pdf(x, alpha, beta, loc=m, scale=1) 
    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name=f'alpha={alpha}'))
```

# Plot

```python
fig.update_layout(
    title="L√©vy Distribution - Different Curves",
    xaxis_title="x",
    yaxis_title="PDF",
    showlegend=True
)
fig.show()
```
